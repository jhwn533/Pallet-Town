```python
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

üîπÎ©îÏÑúÎìú ÌïòÎÇòÎ°ú ÌîåÎ°ØÏùÑ Í∞ÑÎã®Ìûà ÏÉùÏÑ± Ìï† Ïàò ÏûàÍ≥†, Ï∂ïÍ≥º Î≤îÎ°ÄÎèÑ ÏûêÎèôÏúºÎ°ú Ï∂îÍ∞Ä Îê®   
üîπÎç∞Ïù¥ÌÑ∞ ÌäπÏÑ±Ïóê Îî∞Îùº ÌååÎùºÎØ∏ÌÑ∞ Í∏∞Î≥∏Í∞íÏùÑ ÏûêÎèôÏúºÎ°ú ÏÑ†ÌÉù   
   
   ex) ÏÉâÏÉÅ Îß§Ìïë: Îç∞Ïù¥ÌÑ∞ Î≤îÏ£º Íµ¨Î∂ÑÏùÑ ÏúÑÌï¥ Î≥ÄÏàòÎ≥Ñ Í≥†Ïú†ÏÉâÏùÑ ÏûêÎèôÏúºÎ°ú ÏÇ¨Ïö©. Ïà´ÏûêÎ≥ÄÏàòÎäî Ïó∞ÏÜç Í∑∏ÎùºÎç∞Ïù¥ÏÖòÏúºÎ°ú ÌëúÌòÑÎêòÍ∏∞ÎèÑ Ìï®
    

```python
sns.relplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g"
)
```

**Ï∂îÍ∞ÄÏ†ÅÏù∏ Í∑∏ÎûòÌîΩ ÏöîÏÜå ÏÑ§Ï†ï Î∞©Î≤ï**    
        üîπÍ∏∞Î≥∏ ÎÇ¥Ïû• ÌÖåÎßà ÏÇ¨Ïö©   
        üîπÎ©îÏÑúÎìú ÎÇ¥ ÌëúÏ§ÄÌôîÎêú ÌååÎùºÎØ∏ÌÑ∞ ÏÑ§Ï†ï   
        üî∏Ï∂îÍ∞Ä ÌÇ§ÏõåÎìú Ïù∏ÏàòÎ•º Í∏∞Î≥∏ matplotlib artistÏóê Ï†ÑÎã¨ÎêòÏñ¥ Ìõ®Ïî¨ Îçî ÎßéÏùÄ Ï†úÏñ¥ Í∞ÄÎä•   
        üî∏ÌîåÎ°Ø ÏÉùÏÑ± ÌõÑ seaborn APIÏôÄ matplotlib Î†àÏù¥Ïñ¥Î°ú ÎìúÎ°≠ Îã§Ïö¥ÌïòÏó¨ ÏÑ∏Î∞Ä Ï°∞Ï†ïÏúºÎ°ú Ìï¥Îãπ ÏÜçÏÑ± ÏàòÏ†ï Í∞ÄÎä•
        
Ï∞æÏïÑÎ≥º Í≤ÉÎì§ : matplotlib artist, sns API/plt Î†àÏù¥Ïñ¥Î°ú ÎìúÎ°≠Îã§Ïö¥   
* matplotlib artistÏóê Í¥ÄÌïú ÏÑ§Î™Ö : https://gibles-deepmind.tistory.com/35   "ÏïÑÌã∞Ïä§Ìä∏Îäî Ïñ¥ÎñªÍ≤å ÌîåÎ°ØÏùÑ Í∑∏Î¶¥ÏßÄÏóê ÎåÄÌïú Î∂ÄÎ∂Ñ"
* Î¨∏ÏÑú ÎÇ¥Ïö©Ïù¥ Ï†ÑÏ≤¥Ï†ÅÏúºÎ°ú Ïâ¨Ïö¥ seabornÍ≥º Ïñ¥Î†µÏßÄÎßå ÏÑ∏Î∞ÄÌïú Ï°∞Ï†ïÏù¥ Í∞ÄÎä•Ìïú matplotlibÏùÑ ÏÑûÏñ¥ÏÑú Ïì∞Î©¥ Ï¢ãÎã§Îäî ÏñòÍ∏∞Ïù∏ Í±¥ ÏïåÍ≤†ÎäîÎç∞, ÎìúÎ°≠Îã§Ïö¥ÌïúÎã§ÎäîÍ≤å Ï†ïÌôïÌûà Ïñ¥Îñ§ ÏùòÎØºÏßÄ Ïûò Î™®Î•¥Í≤†Îã§.

```python
# ÎÇ¥Ïû• ÌÖåÎßà Ï†ÅÏö©
sns.set_theme(style="ticks", font_scale=1.25)

# ÏÉâÏÉÅ ÌåîÎ†àÌä∏ÏôÄ ÎßàÏª§ÌòïÌÉú ÏÑ§Ï†ï
g = sns.relplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g",
    palette="crest", marker="+", s=100,
)

# Ï∂ï ÎùºÎ≤® ÏÑ§Ï†ï
g.set_axis_labels("Bill length (mm)", "Bill depth (mm)", labelpad=10)

# Ïù∏Îç±Ïä§ ÏÑ§Ï†ï
g.legend.set_title("Body mass (g)")

# ÌîåÎ°Ø ÏÇ¨Ïù¥Ï¶à ÏÑ§Ï†ï
g.figure.set_size_inches(6.5, 4.5)

# Ï∂ïÍ≥ºÏùò Í≥µÍ∞Ñ Î∞è Ï∂ï ÎÅù Î∂ÄÎ∂Ñ ÌòïÌÉú ÏÑ§Ï†ï
g.ax.margins(.15)
g.despine(trim=True)
```
