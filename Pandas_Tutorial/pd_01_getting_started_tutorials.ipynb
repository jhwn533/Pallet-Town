# How do I read and write tabular data?

```python
# íƒ€ì´íƒ€ë‹‰ ìŠ¹ê° ë°ì´í„°(csv) ë¶ˆëŸ¬ì˜¤ê¸°
# read_csv('src') | read_ë°ì´í„°íƒ€ì…(csv, xls, parquet â€¦) ìœ¼ë¡œ ë°ì´í„° ë¡œë“œ
# ë°ì´í„° ë¡œë“œ í›„ í™•ì¸ ìŠµê´€í™”í•˜ê¸°(ê¸°ë³¸ê°’ ì•ë’¤ 5í–‰ì”© ë³´ì—¬ì¤Œ)
titanic = pd.read_csv('data/titanic.csv')
titanic
```
