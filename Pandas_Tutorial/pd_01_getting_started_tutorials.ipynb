# What kind of data does pandas handle?
##### pandas는 어떤 종류의 데이터를 처리하는가?



```python
# 패키지 로드
# pandas의 경우 대부분 관행적으로 pd 라는 별칭으로 로드함
import pandas as pd
```

## pandas data table representation
##### pandas 데이터테이블 표현
![DataFrame](https://pandas.pydata.org/pandas-docs/stable/_images/01_table_dataframe.svg)

```python
# 타이타닉 승객 데이터 저장
# 데이터내용: 이름(characters), 나이(integers), 성별(male/female)
# 수동으로 DataFrame 생성 시, 딕셔너리를 사용하면 키는 열 이름, 값은 열로 저장됨
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
    }
)
```

```python
df
```

`DataFrame`
- 문자, 정수, 소수, 범주형데이터 등 다양한 유형의 데이터를 저장할 수 있는 2차원 데이터구조
- SQL: 테이블 / R: 스프레드 시트, Data.frame과 유사

    `df`
    - 3개의 열(columns)로 구성
    - 열 이름 및 데이터 형태: Name(text), Age(numbers), Sex(text)    

## Each column in a DataFrame is a <span style='color:#BF4545'>Series</span>
##### 데이터프레임의 각 열은 시리즈
![series](https://pandas.pydata.org/pandas-docs/stable/_images/01_table_series.svg)

```python
# 열 선택방법: 대괄호[] 사이에 열 이름 입력(대소문자 구분) - 🏷️ 딕셔너리에서 키로 값을 선택하는 것과 유사
# DataFrame 단일 열은 pandas series 타입
df['Age']
```

```python
# Series 처음부터 만들기
# Series는 단일 열 구조라 열 이름이 없음(행 이름은 있음)
ages = pd.Series([22, 35, 58], name = "Age")
ages
```

## Do something with a DataFrame or Series
##### 데이터프레임 / 시리즈로 작업 수행

```python
# 승객 최대 연령 알아보기
# 데이터프레임에서 > 연령(Age) 열 선택 > 최댓값 함수(max) 적용
df['Age'].max()
```

```python
# 시리즈에서 > 최댓값 함수(max) 적용
# max 외에 다양한 '함수' 적용 가능, 함수이므로 뒤에 괄호 사용
ages.max()
```

```python
# 숫자 데이터 기본 통계 알아보기
# describe()
## 숫자 데이터 개요, 속성 기본값은 숫자데이터만 처리(Name, Sex : 문자데이터) > DataFrame / Series 타입으로 반환
## series에도 적용 가능
## 통계에 사용된 집계방법 대한 상세 정보는 유저 가이드 : describe 참고
df.describe()
```

## summary

1. 패키지 로드 : `import pandas as pd`
2. 데이터 테이블은 pandas `DataFrame` 으로 저장됨
3. 데이터프레임의 각 열은 `Series` 타입임
4. 데이터프레임이나 시리즈에 메서드 적용 가능

# How do I read and write tabular data?
##### 테이블 데이터 읽고 쓰는 법
![table_rw](https://pandas.pydata.org/pandas-docs/stable/_images/02_io_readwrite.svg)

```python
# 타이타닉 승객 데이터(csv) 불러오기
# read_csv('src') | read_데이터타입(csv, xls, parquet …) 으로 데이터 로드
# 데이터 로드 후 확인 습관화하기(기본값 앞뒤 5행씩 보여줌)
titanic = pd.read_csv('data/titanic.csv')
titanic
```

```python
# 일부데이터 불러오기 : head(n) | tail(n)
titanic.head(3)
```

```python
# 열 데이터 속성 확인 : dtypes(소관호 사용 안함 주의 > DataFrame, Series의 '속성'이기 때문)
titanic.dtypes
```

```python
# 데이터 스프레드시트 형태로 내보내기
# to_excel('file name.type') | to_* 데이터 저장 
# sheet_name : 시트명 지정 속성, index : false 설정시 행 인덱스 저장 안함
titanic.to_excel('titanic.xlsx', sheet_name = 'passengers', index = False)
```

```python
pd.read_excel('titanic.xlsx')
```

```python
# 데이터프레임 요약
# info() : 해당 객체 클래스, 크기, 인덱스 범위, 열 개수, 열별 값 개수 및 null 개수, 열별 값 타입(str, object/ integer / float), 사용 램 용량
titanic.info()
```

## summary
1. `read_*` 메서드로 다양한 데이터형식 불러오기
2. `to_*` 메서드로 다양한 데이터형식 내보내기
3. `head()`, `tail()`, `info()` 메서드와 `dtypes` 속성으로 데이터 첫 확인 습관화하기
