# Figure aesthetics
## Controlling figure aesthetics

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def sinplot(n=10, flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, n + 1):
        plt.plot(x, np.sin(x + i * .5) * (n + 2 - i) * flip)
```

```python
sinplot()
```

matplotlib 파라미터를 두 그룹으로 나눔
1. 플롯의 미적 스타일 설정 > 파라미터 딕셔너리 반환  
    `axes_style()`, `set_style()`
2. 플롯 요소들의 크기를 조정해 접근성 높임 > matplotlib 기본값 설정  
    `plotting_context()`, `set_context()`

### `set_style()`, `axes_style()`
- `darkgrid`(default)
- `whitegrid`
- `dark`
- `white`
- `ticks`

```python
f = plt.figure(figsize=(6, 6))
gs = f.add_gridspec(2, 2)

with sns.axes_style("darkgrid"):
    ax = f.add_subplot(gs[0, 0])
    sinplot(6)

with sns.axes_style("white"):
    ax = f.add_subplot(gs[0, 1])
    sinplot(6)

with sns.axes_style("ticks"):
    ax = f.add_subplot(gs[1, 0])
    sinplot(6)

with sns.axes_style("whitegrid"):
    ax = f.add_subplot(gs[1, 1])
    sinplot(6)

f.tight_layout()
```

### `set_context()`
- paper
- notebook(default)
- talk
- poster

```python
f = plt.figure(figsize=(6, 6))
gs = f.add_gridspec(2, 2)

with sns.axes_style("darkgrid"):
    sns.set_context("paper")
    ax = f.add_subplot(gs[0, 0])
    sinplot(6)

with sns.axes_style("darkgrid"):
    sns.set_context("notebook")
    ax = f.add_subplot(gs[0, 1])
    sinplot(6)

with sns.axes_style("darkgrid"):
    sns.set_context("talk")
    ax = f.add_subplot(gs[1, 0])
    sinplot(6)

with sns.axes_style("darkgrid"):
    sns.set_context("poster")
    ax = f.add_subplot(gs[1, 1])
    sinplot(6)

f.tight_layout()
```

## Choosing color palettes
색을 색조(hue), 채도(Saturation(or chroma)), 휘도(lightness) 관점에서 보기
- 색조(hue): '서로 다름'을 구분하는 요소
- 채도(Saturation): 다채로움, 채도가 높을 수록 색상은 더 뚜렷하게 인식
- 휘도(lightness): 검은색에서 흰색까지 빛이 얼마나 많이 방출되는지 차이

색조는 **서로 다른 요소** 표현에 사용, 채도·휘도는 숫자와 같은 **연속적인 요소** 표현에 사용

### color_palette()
- qualitative palettes : 범주형 데이터, 색조 구성 요소의 차이가 큼
- sequential palettes : 수치형 데이터
- diverging palettes : 범주형경계가 있는 수치형 데이터

##### Qualitative color palettes
- ` `(default)
- `tab10`(matplotlib default)
- `pastel`, `muted`, `bright`, `deep`, `colorblind`, `dark`(variations of matplotlib’s palette)
- circular color systems: (기본값에서 사용할 수 있는 색상보다 더 많은 색상이 포함된 범주형 색상표가 필요한 경우)`hls`, `husl`
- categorical Color Brewer palettes: (sequential, diverging에도 포함됨) `set2`, `Paired`



##### Sequential color palettes
- 휘도 중요
- `rocket`, `mako` - 검정에서 흰색까지의 변화, 히트맵 표현 등에 유용
- `flare`, `crest` - 두 색 간의 변화, 도형이나 선 색에 활용
- `magma`, `viridis` - matplotlib
- `_r` - 컬러맵 반전  

이런 순차적 컬러 팔레트는 불연속형 값 생성 가능(극단 값은 사용하지 않음(ex. 검정색, 흰색))

##### Sequential “cubehelix” palettes
- 밝기가 선형적으로 증가/감소하고 색조가 지속적으로 변화하는 팔레트
- `cubehelix` (matplotlib 기본 내장)
- sns.cubehelix_palette()
    - start = 0 ~ 3
    - rot = -1 ~ 1
    - dark, light = 0 ~ 1
    - reverse = T/F

##### Custom sequential palettes
`light_palette()`나 `dark_palette()`로 단일색 연속형 팔레트 생성 가능

##### Sequential Color Brewer palettes

##### Diverging color palettes
- 큰값 작은 값 양쪽 다 중요하고 중간값이 걸쳐있는 데이터, 양쪽 극단에 두개의 주요색상 있음
- `vlag`, `icefire`
- diverging_palette()로 커스텀 생성 가능
- `Spectral`, `coolwarm` (matplotlib 내장)
